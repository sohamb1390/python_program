```
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```
url = 'https://raw.githubusercontent.com/subhajitmajumder/python_program/master/datasets/df1'
df1 = pd.read_csv(url, index_col=0)

url2 = 'https://raw.githubusercontent.com/subhajitmajumder/python_program/master/datasets/df2'
df2 = pd.read_csv(url2)

url3 = 'https://raw.githubusercontent.com/subhajitmajumder/python_program/master/datasets/df3'
df3 = pd.read_csv(url3)
```

We can create a histogram from any dataframe.

```
df1['A'].hist(bins=30)
```

We can also create histogram by using plot method.

```
df1['A'].plot(kind = 'hist', bins=30)
```

Another we can use an other way.

```
df1['A'].plot.hist()
```

```
df2.head()
```

```
df2.plot.area(alpha=0.5)
```

```
df2.plot.bar()
```

```
df2.plot.bar(stacked = True)
```

```
df1.plot.line()
```

```
df1.plot.scatter(x='A', y='B', s=df1['C']*50, cmap='coolwarm')
```

```
df2.plot.box()
```

```
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
```

```
df.plot.hexbin(x='a', y='b', gridsize=25, cmap='coolwarm')
```
