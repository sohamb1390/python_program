```
import matplotlib.pyplot as plt
```

```
#Allows us to see plots inside jupyter notebook if someone use it.
%matplotlib inline 
```

```
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```
x
```

```
y
```

```
#Functional Method of Plotting.
plt.plot(x, y)
```

Note that if we not use that inline command then we need to use **plt.show()** command.Take a look below.This command actually printing the plot instead of showing the plot.

```
plt.plot(x, y)
plt.show()
```

```
#Labelling a plot.
plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('y = x^2')
```

```
#Creating multiplot in same canvas.
plt.subplot(1, 2, 1) # (rows, columns, index)
plt.plot(x, y, 'r') # r tends to red

plt.subplot(1, 2, 2)
plt.plot(y, x, 'g')
```

###Object oriented Method of using matplotlib

```
 fig = plt.figure() #A figure object which is mainly a blank canvas.

 axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #passing a list where it has four arguments which is left, buttom, width and height arguments.

axes.plot(x, y)
axes.set_xlabel('X axis')
axes.set_ylabel('Y axis')
axes.set_title(' Y = X^2')
```

```
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
```

In the list the first value is for left, 2nd value is buttom, 3rd value is width and fourth is height. By plying with the first two values i.e. left and buttom we can shift the small canvas.

```
#Shifting the small canvas within the big canvas.

fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.45, 0.15, 0.4, 0.3])
```

```
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes1.set_title('Larger Plot')

axes2.plot(y, x)
axes2.set_title('Smaller Plot')
```

Creating subplot using object oriented programme.

```
#Two canvases as columns = 2
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout() #It fix all the canvas from overlapping.
```

```
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y) #access any canvas by axes[0], axes[1] and so on
axes[0].set_title('First plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')
```

###Figure Size, Aspect Ratio & DPI

Matplotlib allows us to access aspect ratio of a figure.Lets jump to a new cell.

```
fig = plt.figure(figsize = (3, 2)) # 3 inchs length and 2 inchs breadth

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

We can do it for subplots also.

```
fig, axes = plt.subplots(nrows=2, ncols = 1, figsize = (6, 3))

axes[0].plot(x, y)
axes[1].plot(y, x)
```

###Saving any figure

###Creating legends

```
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label= 'x square')
ax.plot(x, x**3, label= 'x cube')

ax.legend() # here we can use loc argument inside legend bracket fix proper position
```

###Appearance of Plot

```
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color = 'green', linewidth = 3, alpha = 0.9,
                                 linestyle= '-', marker = 'o',
                                 markersize = '10',
                                 markerfacecolor= 'yellow',
                                 markeredgewidth= 3,
                                 markeredgecolor= 'black')

#ax.set_xlim([0, 1])
#ax.set_ylim([0, 2]) 

#Can use custom colors by using hex code
# linewidth parameter is used for width of the line.
# alpha parameter is used for opacity of the line.
#linestyle parameter is used for styling any line.
# marker parameter is used to mark out each point on the graph
```
